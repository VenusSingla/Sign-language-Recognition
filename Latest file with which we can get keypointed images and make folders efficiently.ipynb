```python
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results
```

```python
def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION) # Draw face connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS) # Draw pose connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Draw right hand connections
```

```python
def draw_styled_landmarks(image, results):
    # Draw face connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION, 
                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
                             mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                             ) 
    # Draw pose connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             ) 
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
```

```python
import cv2
import numpy as np
import os

def augment_images(input_folder, output_folder, num_augmented_images=30):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Load the two original images
    image1 = cv2.imread(os.path.join(input_folder, '0.jpg'))
    image2 = cv2.imread(os.path.join(input_folder, '1.jpg'))

    # Resize images to the same dimensions
    image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

    # Apply data augmentation to generate new images
    for i in range(num_augmented_images):
        augmented_image = generate_augmented_image(image1, image2)

        # Save the augmented image
        augmented_image_name = f"augmented_{i}.jpg"
        augmented_image_path = os.path.join(output_folder, augmented_image_name)
        cv2.imwrite(augmented_image_path, augmented_image)

def generate_augmented_image(image1, image2):
    # Choose a random combination of weights for blending the two images
    alpha = np.random.uniform(0.3, 0.7)
    beta = 1 - alpha

    # Blend the two images
    blended_image = cv2.addWeighted(image1, alpha, image2, beta, 0)

    # Apply additional data augmentation if needed (e.g., rotation, scaling, flipping)

    return blended_image

# Example usage:
input_images_folder = "D:/Word_Data/ACCIDENT"
output_images_folder = "D:/Word_Data/ACCIDENT1"
num_augmented_images = 10

augment_images(input_images_folder, output_images_folder, num_augmented_images)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
import os

def augment_keypointed_images(input_folder, output_folder, num_augmented_images=30):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Create an ImageDataGenerator with desired augmentations
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    # Iterate over each folder in the input directory
    for folder in os.listdir(input_folder):
        folder_path = os.path.join(input_folder, folder)

        # Skip non-directory items
        if not os.path.isdir(folder_path):
            continue

        # Create a corresponding folder in the output directory
        output_folder_path = os.path.join(output_folder, folder)
        os.makedirs(output_folder_path, exist_ok=True)

        # Iterate over each keypointed image in the folder
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpg"):
                image_path = os.path.join(folder_path, filename)
                image = cv2.imread(image_path)

                # Reshape the image to comply with Keras input shape
                image = image.reshape((1,) + image.shape)

                # Generate augmented images
                for i, batch in enumerate(datagen.flow(image, batch_size=1, save_to_dir=output_folder_path, save_prefix='aug', save_format='jpg')):
                    if i >= num_augmented_images:
                        break  # Break the loop after generating the desired number of augmented images

# Example usage:
input_images_folder = "D:/Data_1/"
output_images_folder = "D:/Word_Data_Augmented/"
num_augmented_images = 10

augment_keypointed_images(input_images_folder, output_images_folder, num_augmented_images)
```

```python
import os
from PIL import Image
from torch.utils.data import Dataset
from torchvision import transforms, utils
import torch.nn.functional as F
import torch.nn as nn
from tqdm import tqdm
import torch

import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
from datasets import load_dataset
import os
from PIL import Image
from torch.utils.data import Dataset
from torchvision import transforms
class CustomDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = os.listdir(root_dir)
        self.class_to_idx = {cls: i for i, cls in enumerate(self.classes)}
        self.images = self._load_images()
    def _load_images(self):
        images = []
        for class_name in self.classes:
            class_path = os.path.join(self.root_dir, class_name)
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                images.append((img_path, self.class_to_idx[class_name]))
        return images
    def __len__(self):
        return len(self.images)
    def __getitem__(self, idx):
        img_path, label = self.images[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label

# Assuming your generator produces color images
# Define the generator and discriminator networks
class Generator(nn.Module):
    def __init__(self, latent_size, image_size):
        super(Generator, self).__init__()
        # Adjust these values based on your requirements
        channels = 3  # Assuming RGB images
        initial_size = 4  # Initial size before upsampling
        self.model = nn.Sequential(
            nn.Linear(latent_size, 256),
            nn.ReLU(),
            nn.Linear(256, channels * initial_size * initial_size),
            nn.ReLU(),
            nn.Unflatten(1, (channels, initial_size, initial_size)),
            nn.ConvTranspose2d(channels, channels // 2, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(channels // 2, channels // 4, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(channels // 4, channels, kernel_size=4, stride=2, padding=1),
            nn.Tanh()
        )
    def forward(self, x):
        return self.model(x)


class Discriminator(nn.Module):
    def __init__(self, channels, image_size):
        super(Discriminator, self).__init__()

        self.model = nn.Sequential(
            nn.Conv2d(channels, 32, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
        )

        # Calculate the size of the output from the convolutional layers
        conv_output_size = self._calculate_conv_output_size(channels, image_size)

        self.fc = nn.Sequential(
            nn.Linear(64 * conv_output_size * conv_output_size, 128),
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def _calculate_conv_output_size(self, channels, image_size):
        dummy_input = torch.rand(1, channels, image_size, image_size)
        dummy_output = self.model(dummy_input)
        conv_output_size = dummy_output.view(1, -1).size(1)
        return conv_output_size

    def forward(self, x):
        x = self.model(x)
        x = x.view(x.size(0), -1)  # Flatten the output before passing it to the fully connected layer
        return self.fc(x)

# Assuming your generator produces color images
channels = 3  
latent_size = 50
image_size = 32 
batch_size = 8
epochs = 4
learning_rate = 0.0002


# Initialize generator and discriminator
generator = Generator(latent_size, image_size)

discriminator = Discriminator(channels, image_size)
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    # Add other transformations as needed
])



dataset = CustomDataset(root_dir='D:/Data_1/', transform=transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
# Loss function and optimizers
criterion = nn.BCELoss()
optimizer_g = optim.Adam(generator.parameters(), lr=learning_rate)
optimizer_d = optim.Adam(discriminator.parameters(), lr=learning_rate)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
generator.to(device)
discriminator.to(device)

# Training loop
for epoch in range(epochs):
    for real_images, _ in tqdm(dataloader):
        batch_size = real_images.size(0)

        # Train discriminator
        real_labels = torch.ones(batch_size, 1)
        fake_labels = torch.zeros(batch_size, 1)

        optimizer_d.zero_grad()

        # Forward pass with real images
        outputs_real = discriminator(real_images)
        d_loss_real = criterion(outputs_real, real_labels)
        d_loss_real.backward()

        # Forward pass with fake images
        z = torch.randn(batch_size, latent_size)
        fake_images = generator(z)
        outputs_fake = discriminator(fake_images.detach())
        d_loss_fake = criterion(outputs_fake, fake_labels)
        d_loss_fake.backward()

        optimizer_d.step()

        # Train generator
        optimizer_g.zero_grad()

        # Forward pass with fake images (no detach for backpropagation)
        outputs_fake = discriminator(fake_images)
        g_loss = criterion(outputs_fake, real_labels)
        g_loss.backward()

        optimizer_g.step()


# Save and visualize generated images in respective directories
for i in range(num_generated_images):
    z_new = torch.randn(1, latent_size)
    generated_image = generator(z_new)

    # Convert generated image tensor to PIL image
    generated_image_pil = transforms.ToPILImage()(generated_image.squeeze())

    # Get the class label (modify this based on your actual data)
    class_label = 0  # Replace with the actual class label

    # Create directory for the class label if it doesn't exist
    class_dir = os.path.join('D:/Word_Data_GAN/', f'class_{class_label}')
    os.makedirs(class_dir, exist_ok=True)

    # Save the image
    save_path = os.path.join(class_dir, f'generated_image_{i + 1}.png')
    generated_image_pil.save(save_path)

    # Visualize the image (optional)
    if generated_image.shape[0] == 1:
        plt.imshow(generated_image_pil, cmap='gray')
    else:
        plt.imshow(generated_image_pil)
    
    plt.title(f'Generated Image {i + 1}')
    plt.show()
```
